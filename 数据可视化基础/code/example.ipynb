## 离散数据与连续数据

### 连续数据采样

```python
import numpy as np

X = np.arange(0, 10, 1)
print(X)
```

```python
Y = np.sin(X)
print(Y)
```

```python
import matplotlib.pyplot as plt
plt.plot(X,Y)
plt.savefig('discret')
plt.show()
```

```python
X = np.arange(0, 10, 0.5)
Y = np.sin(X)
plt.plot(X,Y)
plt.show()
```

```python
X = np.arange(0, 10, 0.1)
Y = np.sin(X)
plt.plot(X,Y)
plt.savefig('continue')
plt.show()
```

## 文本数据可视化

### 数据导入

```python
with open('danmu_12-1-13.csv', encoding='utf8') as file:
    content = file.read()
print(content)
```

### 数据清洗

```python
import jieba
```

```python
content = content.strip().replace("\n", "")
```

```python
content[:1000]
```

#### 分词

```python
words = jieba.lcut(content)
```

```python
print(words[:100])
```

#### 去除停用词

```python
stopwords = []
with open('stopwords-master/cn_stopwords.txt', encoding='utf8') as file:
    for line in file:
        word = line.strip()
        stopwords.append(word)
with open('stopwords-master/baidu_stopwords.txt', encoding='utf8') as file:
    for line in file:
        word = line.strip()
        stopwords.append(word)
with open('stopwords-master/hit_stopwords.txt', encoding='utf8') as file:
    for line in file:
        word = line.strip()
        stopwords.append(word)
with open('stopwords-master/scu_stopwords.txt', encoding='utf8') as file:
    for line in file:
        word = line.strip()
        stopwords.append(word)


print(stopwords[:10])
```

```python
print(len(words))
```

```python
words = [word for word in words if word not in stopwords]
```

```python
print(len(words))
```

### 数据可视化

```python
import wordcloud
```

```python
wc = wordcloud.WordCloud(font_path="msyh.ttc",max_words=30, background_color="white") 
wc.generate(' '.join(words))
wc.to_file('wordcloud.jpg')
```
