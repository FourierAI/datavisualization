### 等高线图 

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
Levels = 8
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
Z = f(X, Y) 
C = plt.contour(X,Y, Z, Levels)
plt.contourf(X,Y, Z, Levels, alpha = 0.75, cmap = plt.cm.hot)

# 添加图例
plt.clabel(C, inline = True)
plt.show()
```

### 饼状图 pie

```python
x = [0.7, 0.3]
plt.pie(x, labels=["A","B"], autopct='%.2f%%' )
plt.show()
```

### 直方图 hist

```python
x = np.random.randn(10000)
plt.hist(x, bins= 1000, density = True)
plt.show()
```

### 箱线图 boxplot

```python
x1 = np.random.normal(0, 2, 100)
x2 = np.random.normal(5, 2, 100)
x3 = np.random.normal(6, 3, 100)
plt.boxplot([x1,x2,x3],labels = ['x1','x2','x3'])
plt.grid()
plt.show()
```

### 三维图 plot_surface

```python
from mpl_toolkits.mplot3d import Axes3D
# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig)
surf = ax.plot_surface(X,Y,Z, cmap=plt.get_cmap('rainbow'))
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
